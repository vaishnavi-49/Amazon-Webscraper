```python
import selenium
from selenium import webdriver
import pandas as pd
import time
```

```python
driver = webdriver.Chrome()
driver.get('https://www.amazon.in')
driver.maximize_window()
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

user_item_input = input("Search an item from amazon")
print(f"{user_item_input} entered!")

search_box=driver.find_element(By.ID,'twotabsearchtextbox')
search_box.clear()
search_box.send_keys(user_item_input)

driver.find_element(By.ID,'nav-search-submit-button').click()
```

```python
driver.find_element(By.XPATH,'//*[@id="brandsRefinements"]/ul/span/li/span/div/a/span').click()

brands_element = driver.find_element(By.XPATH, '//*[@id="brandsRefinements"]/ul')
# print(brands_element.text)
brands = brands_element.text.split('\n')
brands.pop()
print("Select a brand from the below list")
for brand in brands:
    print(brand)
```

```python
# Prompt the user to enter some text
user_input = input("Enter the brand: ")

#Enter the brand and laptop in search box
search_box=driver.find_element(By.ID,'twotabsearchtextbox')
search_box.clear()
search_box.send_keys(user_input + " " + user_item_input)

driver.find_element(By.ID,'nav-search-submit-button').click()
try:
    driver.find_element(By.XPATH,"//span[text()='"+ user_input +"']").click()
except Exception as e:
    print(f"Brand not found")
```

```python
laptop_names=[]
laptop_actual_prices=[]
laptop_prices=[]
laptop_discounts=[]
laptop_reviews=[]
laptop_review_ratings=[]
laptop_links=[]

# Add a delay to allow the page to load
time.sleep(3)
page_count = 0;
while (page_count<10):
    try:
        names=driver.find_elements(By.XPATH,"//span[@class='a-size-medium a-color-base a-text-normal']")
        for name in names:
            laptop_names.append(name.text)
            
        actual_prices=driver.find_elements(By.XPATH,"//span[@class='a-price a-text-price']")
        for actual_price in actual_prices:
            laptop_actual_prices.append(actual_price.text)
            
        prices=driver.find_elements(By.XPATH,"//span[@class='a-price-whole']")
        for price in prices:
            laptop_prices.append(price.text)
            
        discounts=driver.find_elements(By.XPATH,"//span[@class='a-truncate-cut']")
        for discount in discounts:
            laptop_discounts.append(discount.text)
        
            
        reviews=driver.find_elements(By.XPATH,"//span[@class='a-size-base s-underline-text']")
        for review in reviews:
            laptop_reviews.append(review.text)
        
        # Find the element using the relevant selector 
        rating_elements = driver.find_elements(By.CSS_SELECTOR, 'span[aria-label*="out of 5 stars"]') 
        # review_ratings=driver.find_elements(By.XPATH,"//span[@class='a-price-whole']")
        for rating_element in rating_elements:
            laptop_review_ratings.append(rating_element.get_attribute('aria-label')[:3])

        product_links = driver.find_elements(By.XPATH, "//a[@class='a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal']")
        for product_link in product_links:
            laptop_links.append(product_link.get_attribute('href'))
            
     
        # Find the "Next" button using XPATH or CSS selector
        next_button = driver.find_element(By.XPATH, "//a[contains(text(), 'Next')]")
        # Check if the "Next" button is disabled
        is_disabled = next_button.get_attribute("aria-disabled") == "true"
        if is_disabled:
            print("The Next button is disabled. Stopping the pagination.")
            break  # Exit the loop if the button is disabled
        # Click the "Next" button
        next_button.click()  
        page_count=page_count+1
        # Add a delay to allow the new page to load
        time.sleep(3)  # Adjust the sleep time as needed
    except Exception as e:
        print(f"An error occurred")
        break  # Exit the loop in case of an error (e.g., element not found)
        
df=pd.DataFrame(zip(laptop_names,laptop_actual_prices,laptop_prices,laptop_reviews,laptop_review_ratings,laptop_links ),columns=['Description','Actual_price','Discounted_price','Reviews', 'Rating', 'Link'])
df
```

```python
# Convert "Actual_price" and "Discounted_price" columns to numeric by removing any currency symbols and commas
df['Actual_price'] = df['Actual_price'].replace('[\₹,]', '', regex=True).astype(float)
df['Discounted_price'] = df['Discounted_price'].replace('[\₹,]', '', regex=True).astype(float)

# Calculate the discount percentage
df['Discount %'] = (((df['Actual_price'] - df['Discounted_price']) / df['Actual_price']) * 100).astype(int)

# Display the dataframe
df
```
